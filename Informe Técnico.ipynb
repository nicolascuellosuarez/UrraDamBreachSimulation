## **Ahora bien, ¿Cómo Escribimos la Dinámica Acuática?**

Vamos a hacer un **ejercicio mental**. Imaginemos que tenemos una fotografía **aérea** del valle del **río Sinú**, justo donde está la represa de **Urrá**. En esa fotografía, podemos **ver el embalse**, **la cortina de concreto**, y aguas abajo el cauce del río serpenteando entre montañas. Ahora, imaginemos que nos piden describir, **con números**, lo que pasaría si esa cortina desapareciera de golpe. **¿Por dónde empezamos?** Necesitamos saber, inicialmente, qué tanta agua hay y hacia dónde y qué tan rápido se está moviendo en cada punto de la **fotografía**. No basta con decir **"el agua va hacia el norte"**, por ejemplo, pues en un punto, el agua puede ir **rápido** hacia el norte; en otro, puede estar **remansada**; en otro, puede estar formando un remolino. La descripción completa del fenómeno **requiere** tres números **en cada punto del valle**: 

- La profundidad del agua, llamémosla $h$.
- La velocidad en la dirección este-oeste **(velocidad horizontal)**, llamémosla $u$.
- La velocidad en la dirección norte-sur **(velocidad vertical)**, denotémosla $v$. 

Y estos tres números **cambian** con cada unidad de tiempo que pasa, además, también **cambian** si nos desplazamos de un lugar a otro. Por eso, en el **lenguaje de la física** y las **matemáticas**, decimos que $h$, $u$ y $v$ son **campos escalares**: funciones que toman un punto del espacio y un instante de tiempo, y devuelven un número.

Ahora, **¿qué hace que estos campos cambien?** **¿Qué leyes físicas gobiernan si el agua se acelera, se frena, se acumula o se vacía?** La respuesta viene de dos principios muy básicos en términos **matemáticos**:

El primero es la **conservación de la masa**. Este principio nos dice que el agua **(la masa)** **no se crea ni se destruye**, si en una región del valle está **entrando** más agua de la que **sale**, el nivel tiene que **subir**. Si está **saliendo** más de la que **entra**, el nivel tiene que **bajar**. Esto es tan **intuitivo** como llenar un balde con una **manguera**: si el chorro que **entra** es más potente que lo que se **escapa** por abajo, el agua **sube**; si la fuga es mayor, el **agua baja**. Matemáticamente, esta idea se **escribe** como una ecuación diferencial que relaciona el cambio de una función multivariable como **la profundidad**, con respecto a una variable como **el tiempo**, es decir: 

$$\frac{\partial h}{\partial t}$$

Con cómo varían los **caudales** en el espacio, es decir: 

$$\frac{\partial(hu)}{\partial x}, \frac{\partial(hv)}{\partial y}$$

Tal que la ecuación es:

$$\frac{\partial h}{\partial t} + \frac{\partial(hu)}{\partial x} + \frac{\partial(hv)}{\partial y} = 0$$

El término $\frac{\partial h}{\partial t}$ nos dice qué tan rápido está subiendo o bajando el agua en un punto fijo. Si es positivo, **el agua está subiendo**; si es negativo, **está bajando**. Los términos $\frac{\partial(hu)}{\partial x}$ y $\frac{\partial(hv)}{\partial y}$ nos dicen cómo están cambiando los **caudales** al movernos hacia el **este** y hacia el **norte**. Si al movernos hacia el este el caudal $hu$ aumenta, significa que está saliendo más agua por la derecha de la que entra por la izquierda, y entonces la profundidad tiende a **disminuir**. Esa es la esencia de la conservación de la masa: *lo que entra menos lo que sale es igual a lo que se acumula*.

El **segundo principio** es la **conservación de la cantidad de movimiento**, que es una forma elegante de decir que el agua se mueve porque hay fuerzas que la empujan, y que su velocidad cambia en respuesta a esas fuerzas. Ahora, **¿qué fuerzas actúan sobre una partícula de agua en un río?** 

Principalmente dos. La primera es la **gravedad**, pero no de cualquier manera: lo que **empuja** el agua no es la gravedad en sí misma, sino la **pendiente** de la **superficie** del agua. Imaginemos una tabla inclinada: si ponemos una canica en la parte alta, **rueda hacia abajo**. Cuanto más inclinada está la tabla, más rápido rueda. Lo mismo pasa con el agua: si la superficie del agua está **más alta** atrás que adelante, el agua siente un **empuje hacia adelante**. Ese empuje es proporcional a qué tan inclinada está esa superficie, es decir, a $\frac{\partial(h + zb)}{\partial x}$ **(la pendiente de la superficie libre en dirección este)** y a $\frac{\partial(h + zb)}{\partial y}$ **(la pendiente en dirección norte)**.

La segunda fuerza es la **fricción** con el fondo; el lecho del río **no es liso**; tiene *piedras*, *arena*, *vegetación*. Todo eso **frena** el agua. Cuanto **más rápido** va el agua, **más la frena el fondo**. En la literatura,según lo investigado, se usa un **coeficiente empírico**, el cual es conocido como **el coeficiente de Manning** (lo llamaremos $n$), para medir qué tan rugoso es el fondo. Cuanto más grande es $n$, más fricción hay y más se **frena** el agua.

Ahora, teniendo en cuenta todo lo que se **desarrolló anteriormente** en términos físicos para la ecuación de mecánica hídrica, podemos condensar todas nuestras ideas en una **ecuación**, y al hacerlo,legamos a lo que se conoce como las ecuaciones de **Aguas Someras** o ecuaciones de ***Saint-Venant***. No son ecuaciones sencillas, pero, mirémoslas con calma; para el modelado de la dirección **este - oeste**, la ecuación que se plantea es:

$$\frac{\partial(hu)}{\partial t} + \frac{\partial}{\partial x} \left(hu² + \frac{gh²}{2}\right) + \frac{\partial(huv)}{\partial y} = -gh \frac{\partial z_{b}}{\partial x} - \frac{\tau_{bx}}{\rho}$$

Y para la otra dirección, norte - sur, tenemos:

$$\frac{\partial(hv)}{\partial t} + \frac{\partial(huv)}{\partial x} + \frac{\partial}{\partial y}\left(hv² + \frac{gh²}{2}\right) = -gh \frac{\partial z_{b}}{\partial y} - \frac{\tau_{by}}{\rho}$$



¡Excelente!, Tenemos las ecuaciones a modelar, pero... **¿qué significan?** No nos asustemos por la cantidad de símbolos, vamos a analizarlas **pieza por pieza**: En la primera, el término $\frac{\partial(hv)}{\partial t}$ describe el cambio del caudal *(profundidad por velocidad)* con respecto al **tiempo**. SI este término es grande, el caudal está **aumentando** en ese punto en específico; los términos que le siguen, como $\frac{\partial}{\partial x} \left(hu² + \frac{gh²}{2}\right)$, describen el transporte de la **velocidad** con el movimiento del agua: es la manera en que el agua que **está detrás**, empuja a la que está adelante. El lado derecho de la **ecuación** contiene las fuerzas: $-gh \frac{\partial z_{b}}{\partial x}$, que es la fuerza de la gravedad, que se debe a la pendiente del fondo **(si baja a la derecha, es negativo, por lo que el signo menos de delante lo vuelve positivo, y el agua acelera a la derecha)**. Finalmente $ - \frac{\tau_{bx}}{\rho}$ es la fuerza de fricción, que se opone siempre a la dirección del movimiento.

En conjunto, estas **tres ecuaciones** (una para la masa y dos para la cantidad de movimiento) nos dicen, en el lenguaje más preciso que tenemos, cómo se comporta el agua cuando fluye por **un valle**. Sin embargo, debemos hacer una precisión importante, la complejidad para la **resolución** de expresiones de este tipo, es **extremadamente alta**, y en muchos casos irresoluble, puesto que muchas veces, las funciones encontradas no pueden expresarse mediante **funciones elementales** del cálculo. 

Debido a su inherente y elevada **complejidad**, las ecuaciones vistas **anteriormente** no son **igualdades** que se puedan resolver con **lápiz** y **papel**, a menos que se estén tratando los casos más simples. Para un **valle** real, como el del **río Sinú** con montañas y valles, la única manera de obtener predicciones de la dinámica, es usar un método numérico que permita la discretización de estas ecuaciones para su resolución, sin embargo,

## **Nuestra Resolución Numérica: La Computación Como Vehículo de Resolución**

La respuesta a las anteriores **preguntas** es que sí, hay una infinidad de *"Solvers"* elaborados que permiten una resolución computacional óptima para este problema, y existe, **en específico**, un método numérico efectivo para la solución de las anteriores **ecuaciones diferenciales**. La idea del método es tan antígua como el **mismo cálculo**, pero su implementación práctica solo fue posible luego de la llegada de las **máquinas computacionales**.

Consiste en, básicamente, dividir el **dominio** en una ***malla de pequeñas cceldas***; digamos que cada celda tiene $\Delta x$ metros de ancho y $\Delta y$ metros de alto, y también convengamos dividir el tiempo en pequeños paso de duración $\Delta t$. Así pues, en cada celda, almacenamos los valores de $h$, $u$ y $v$ en el centro de la celda, para posteriormente, en cada paso de **tiempo**, usamos los valores de las **celdas vecinas** para aproximar las **derivadas parciales** que aparecen en las ecuaciones.

Por ejemplo, la derivada $\frac{\partial(hu)}{\partial x}$  en una celda se puede **aproximar** como la *diferencia* entre el **caudal** que sale por la **cara derecha** y el caudal que entra por la **cara izquierda**, dividido por $\Delta x$. Esto convierte las *ecuaciones diferenciales* en *ecuaciones algebráicas* que una computadora puede resolver paso a paso.

Ahora bien, **¿cómo implementamos lo anterior entonces?**, **¿Qué herramientas computacionales utilizamos?**, **¿Cómo?**

Para cumplir con el **objetivo central** de este proyecto, el de generar una abstracción computacional de este fenómeno, implementamos un *"solver"* para estas ecuaciones, utilizando el lenguaje de programación **C**, en el que se recorre la maya celda por celda, calcula los nuevos valores de $h$, $u$ y $v$ a partir de los valores actuales, y luego avanza el tiempo un paso $\Delta t$, luego, se repite este proceso hasta que el reloj marca el **tiempo total** que queremos simular; este es un problema de gran demanda, incluso para un computador, pues, como dijimos antes, es un **volúmen enorme** de operaciones, e incluso una máquina computacional, creada para manejar esta gran cantidad de cálculos, la demanda es enorme, por lo que, precisamente para este **tipo de problemas** es que utilizamos una herramienta central en este informe: **La Computación de Alto Desempeño**. 

Para simular el cambio en cada una de las celdas de la malla reiteradamente, se necesitan **decenas de miles** de pasos de tiempo, lo que se traduce en cientos de miles de operaciones, para **volúmenes** tan grandes de trabajo, lo más óptimo computacionalmente hablando, es paralelizar estos cálculos, dividirlos en **diferentes hilos** de ejecución computacional, para así obtener los resultados de una forma más **eficiente**. Esta fue la razón por la que implementamos los cálculos discretos con directivas de paralelización de **OpenMP**, para que múltiples núcleos del procesador trabajen al mismo tiempo en diferentes regiones de la malla.

Observemos pues, el código realizado para ejecutar las simulaciones:

````c
#include <stdio.h>
#include <stdlib.h>
#include <math.h>
#include <omp.h>
#include "init.h"
#include "utils.h"

// Domain
#define NX 600
#define NY 520
#define DX 50.0
#define DY 50.0
#define TOTAL_TIME 600.0  

// Physics
#define G 9.81
#define MANNING_N 0.035

// Urrá Conditions
#define H_UPSTREAM 130.5
#define H_DOWNSTREAM 68.0
#define DAM_X_POS 314
#define DAM_HEIGHT 73.0
#define SLOPE 0.00015

// Breaking of Dam
#define BREACH_WIDTH 200.0
#define BREACH_START_TIME 5.0
#define BREACH_FORMATION_TIME 60.0

// Control
#define USE_REAL_TOPOGRAPHY 1
#define CFL_NUMBER 0.08  

double calculate_dt_adaptive(double *h, double *u, double *v, int nx, int ny, double dx, double cfl);

int main() {
    printf("\n========================================\n");
    printf("SIMULACIÓN ROTURA DE PRESA - URRÁ\n");
    printf("========================================\n\n");
    
    printf("Configuración del dominio:\n");
    printf("  Distancia total: %.1f km\n", (NX * DX) / 1000.0);
    printf("  Ancho del valle: %.1f km\n", (NY * DY) / 1000.0);
    printf("  Posición de la presa: celda %d (%.0f m)\n", DAM_X_POS, DAM_X_POS * DX);
    printf("  Resolución: %.0f x %.0f metros por celda\n", DX, DY);
    printf("  Total de celdas: %d (%.2f millones)\n\n", NX * NY, (NX * NY) / 1e6);

    int total_cells = NX * NY;
    double *h = (double*)malloc(total_cells * sizeof(double));
    double *u = (double*)malloc(total_cells * sizeof(double));
    double *v = (double*)malloc(total_cells * sizeof(double));
    double *h_new = (double*)malloc(total_cells * sizeof(double));
    double *u_new = (double*)malloc(total_cells * sizeof(double));
    double *v_new = (double*)malloc(total_cells * sizeof(double));
    double *zb = (double*)malloc(total_cells * sizeof(double));

    if (!h || !u || !v || !h_new || !u_new || !v_new || !zb) {
        printf("Error reservando memoria.\n");
        return -1;
    }

    if (USE_REAL_TOPOGRAPHY) {
        printf("Cargando topografía real...\n");
        charge_real_topography(zb, NX, NY, "/home/nicolas/UrraDamBreachSimulation/data/urra_specific_topography.bin");
    } else {
        printf("Usando topografía sintética\n");
    }

    printf("Inicializando condiciones de la presa.\n");
    initialize_simulation_parameters(h, u, v, zb, NX, NY, DX, DY, 
                                    H_UPSTREAM, H_DOWNSTREAM, 
                                    DAM_X_POS, DAM_HEIGHT, 
                                    SLOPE, USE_REAL_TOPOGRAPHY);
    
    double initial_volume = calculate_volume(h, total_cells, DX, DY);
    printf("\nVolumen inicial: %.3f millones de m³\n", initial_volume / 1e6);

    printf("\nIniciando Simulación con CFL adaptativo...\n");
    double start_time = omp_get_wtime();

    double current_time = 0.0;
    int step = 0;
    
    while (current_time < TOTAL_TIME) {
        double dt = calculate_dt_adaptive(h, u, v, NX, NY, DX, CFL_NUMBER);
        if (dt < 0.001) dt = 0.001;
        if (current_time + dt > TOTAL_TIME) dt = TOTAL_TIME - current_time;
        
        apply_breach(h, u, v, zb, NX, NY, DX, DY, 
                    current_time, DAM_X_POS, BREACH_WIDTH, 
                    BREACH_START_TIME, BREACH_FORMATION_TIME);
        
        solver_pass(h, u, v, h_new, u_new, v_new, zb,
                    NX, NY, DX, DY, dt, MANNING_N);
        
        #pragma omp parallel for
        for (int i = 0; i < total_cells; i++) {
            h[i] = h_new[i];
            u[i] = u_new[i];
            v[i] = v_new[i];
        }
        
        current_time += dt;
        step++;
        
        static double last_save_time = 0;
        if (current_time - last_save_time >= 30.0 || step == 1) {
            save_snapshot(h, step, NX, NY, "/home/nicolas/UrraDamBreachSimulation/snapshots");
            last_save_time = current_time;
            
            double max_h = 0.0, min_h = 1e9, max_u = 0.0, max_v = 0.0;
            for (int i = 0; i < total_cells; i++) {
                if (h[i] > max_h) max_h = h[i];
                if (h[i] < min_h) min_h = h[i];
                if (fabs(u[i]) > max_u) max_u = fabs(u[i]);
                if (fabs(v[i]) > max_v) max_v = fabs(v[i]);
            }
            printf("T=%.1fs | paso=%d | dt=%.4f | h=[%.4f,%.2f] | u_max=%.2f | v_max=%.2f\n", 
                   current_time, step, dt, min_h, max_h, max_u, max_v);
        }
        
        if (step % 500 == 0) {
            int error = 0;
            for (int i = 0; i < total_cells && !error; i++) {
                if (isnan(h[i]) || isnan(u[i]) || isnan(v[i])) {
                    printf("ERROR: NaN en paso %d, celda %d\n", step, i);
                    error = 1;
                }
            }
            if (error) break;
        }
    }

    double end_time = omp_get_wtime();
    printf("\n========================================\n");
    printf("RESUMEN FINAL\n");
    printf("========================================\n");
    printf("Simulación completada en %.2f segundos\n", end_time - start_time);
    printf("Pasos totales: %d\n", step);
    printf("Tiempo simulado: %.1f segundos (%.1f minutos)\n", current_time, current_time/60.0);
    
    double final_volume = calculate_volume(h, total_cells, DX, DY);
    printf("\nVolumen inicial: %.3f millones m³\n", initial_volume / 1e6);
    printf("Volumen final:   %.3f millones m³\n", final_volume / 1e6);
    printf("Diferencia: %.3f millones m³ (%.1f%%)\n", 
           (initial_volume - final_volume) / 1e6,
           (initial_volume - final_volume) / initial_volume * 100);
    
    save_snapshot(h, step, NX, NY, "snapshots");

    free(h); free(u); free(v); free(h_new); 
    free(u_new); free(v_new); free(zb);

    printf("\nSimulación finalizada.\n");
    return 0;
}

double calculate_dt_adaptive(double *h, double *u, double *v, int nx, int ny, double dx, double cfl) {
    double max_speed = 0.0;
    
    #pragma omp parallel for reduction(max:max_speed)
    for (int i = 0; i < nx * ny; i++) {
        double c = sqrt(G * fmax(h[i], 0.05));
        double vel = sqrt(u[i]*u[i] + v[i]*v[i]);
        double wave_speed = c + vel;
        if (wave_speed > max_speed) max_speed = wave_speed;
    }
    
    double dt = cfl * dx / (max_speed + 0.001);
    return fmin(dt, 0.5);
}
````

Ahora bien, en nuestro **código** de ***ejecución de simulación***, lo primero que hacemos es **incluir** las cabeceras necesarias. `stdio.h` nos permite usar `printf` para mostrar información en pantalla, `stdlib.h` nos da `malloc` y `free` para manejar la **memoria dinámica**, `math.h` nos permite usar funciones **matemáticas** como `sqrt`, `pow`, `fabs` e `isnan`, y `omp.h` es la cabecera de **OpenMP**, que nos da funciones como `omp_get_wtime` para medir tiempos y las directivas que le dicen al compilador cómo **paralelizar** nuestro código. También incluimos nuestros propios archivos `init.h` y `utils.h`, donde están declaradas las **funciones** que definimos en otros archivos **(que posteriormente serán explicadas)**, como la inicialización y las funciones auxiliares.

Luego definimos una ***serie de constantes** con `#define`, esto es útil porque si queremos **cambiar algún parámetro**, solo lo modificamos en un solo lugar y todo el programa se ajusta automáticamente. Las **constantes del dominio** son `NX` y `NY`, que son la cantidad de celdas en las direcciones $x$ e $y$, y las elegimos como 600 y 520 respectivamente porque queremos que el dominio tenga **30 kilómetros de largo** (600 por 50 metros) y **26 kilómetros de ancho** (520 por 50 metros). `DX` y `DY` son el tamaño de cada celda en metros, y los fijamos en **50 metros** como un buen equilibrio entre precisión y costo computacional. `TOTAL_TIME` es el tiempo total que vamos a simular en segundos.

Las constantes físicas son `G`, la **aceleración de la gravedad**, que vale $9.81$ metros por segundo cuadrado, y `MANNING_N`, el ***coeficiente de rugosidad de Manning*** mencionado anteriormente, que mide la fricción que opone el lecho del río al flujo del agua. Elegimos 0.035, pues investigamos y es un **valor típico** para un río con fondo de grava y **algo de vegetación**.

Las **condiciones específicas** de la presa de Urrá son `H_UPSTREAM` y `H_DOWNSTREAM`, que representan los niveles del agua **aguas arriba** y **aguas abajo**, 130.5 y 68.0 metros sobre el nivel del mar respectivamente. `DAM_X_POS` es la **posición de la presa dentro de la malla**, y la calculamos con las coordenadas reales de la cortina: está en la celda 314, que corresponde a 15.7 kilómetros desde el borde izquierdo del dominio. `DAM_HEIGHT` es la **altura física** de la presa, 73 metros, y `SLOPE` es la pendiente promedio del río aguas abajo, un número muy chico porque el Sinú baja muy suavemente hacia el mar.

Los parámetros de la rotura son `BREACH_WIDTH`, el ancho de la brecha que se abre en la presa, que fijamos en **200 metros**; `BREACH_START_TIME`, el tiempo en segundos después del cual comienza a formarse la brecha, que pusimos en 5 segundos para que la simulación arranque con la presa intacta; y `BREACH_FORMATION_TIME`, que es el tiempo que tarda la brecha en formarse completamente, 60 segundos, durante el cual la **abertura crece gradualmente**. Finalmente, **dos constantes de control**: `USE_REAL_TOPOGRAPHY` es un interruptor que cuando vale 1 le dice al programa que cargue la topografía real del DEM de Urrá, y cuando vale 0 que use una topografía sintética con pendiente suave; y `CFL_NUMBER` es el factor de ***Courant-Friedrichs-Lewy***, que controla la estabilidad del método numérico, y lo fijamos en 0.08, un valor conservador que hace el método más lento pero más estable.

Después de las constantes, declaramos el prototipo de la función `calculate_dt_adaptive`, que calcula el paso de tiempo adaptativo, y luego comenzamos con la función `main`. Dentro del `main`, lo primero que hacemos es mostrar en pantalla la configuración del dominio: la distancia total en kilómetros, el ancho del valle, la posición de la presa en metros, la resolución de la malla, y la cantidad total de celdas. Esto es útil para que el usuario pueda verificar los parámetros sin tener que leer el código.

Luego, reservamos memoria para los arreglos que vamos a usar. Necesitamos la profundidad actual `h`, las velocidades actuales `u` y `v`, y luego necesitamos los arreglos para la profundidad nueva `h_new`, las velocidades nuevas `u_new` y `v_new`, donde almacenaremos los valores del siguiente paso de tiempo sin pisar los actuales. También reservamos memoria para la topografía `zb`. La cantidad total de celdas es `NX * NY`, que es 312,000. Si por alguna razón no se puede reservar la memoria (si `malloc` devuelve `NULL`), mostramos un error y salimos del programa. Después de reservar memoria, cargamos la topografía. Si `USE_REAL_TOPOGRAPHY` es 1, llamamos a la función `charge_real_topography`, que abre un archivo binario con la topografía real de Urrá y lo lee completo dentro del arreglo `zb`. Si el archivo no existe, la función muestra un error. Si `USE_REAL_TOPOGRAPHY` es 0, entonces más adelante, en la función de inicialización, se generará una topografía sintética.

Luego llamamos a la función `initialize_simulation_parameters`. Esta función recibe los arreglos `h`, `u`, `v`, `zb`, las dimensiones de la malla, el tamaño de las celdas, los niveles del agua aguas arriba y aguas abajo, la posición de la presa, su altura, la pendiente y el flag de topografía real. Lo que hace es llenar los arreglos con los valores iniciales: la profundidad `h` se calcula como `H_UPSTREAM - zb` en las celdas que están a la izquierda de la presa, y como `H_DOWNSTREAM - zb` en las celdas que están a la derecha. Las velocidades `u` y `v` se inicializan en cero, porque al principio el agua está quieta. Una vez que tenemos todo inicializado, calculamos el volumen inicial de agua con la función `calculate_volume`. Esta función recorre todas las celdas, suma las profundidades y multiplica por el área de cada celda, que es `DX * DY`. Mostramos este número en pantalla, porque es importante para después compararlo con el volumen final y medir qué tan bien se conserva la masa.

Ahora comienza el bucle temporal, que se ejecuta mientras el tiempo actual no haya alcanzado `TOTAL_TIME`. Dentro del bucle, primero calculamos el paso de tiempo llamando a `calculate_dt_adaptive`. Esta función recibe los arreglos `h`, `u`, `v`, las dimensiones de la malla, el tamaño de celda `DX` y el factor `CFL_NUMBER`. Lo que hace es recorrer todas las celdas y encontrar la velocidad máxima de onda, que es la suma de la velocidad de una onda superficial 

$$c = \sqrt{g \cdot h}$$

Más la velocidad del flujo. Luego calcula `dt` como `CFL_NUMBER * DX / velocidad_maxima`. Si `dt` es menor a 0.001 segundos, lo llevamos a ese mínimo para evitar pasos absurdamente pequeños, y si el tiempo actual más `dt` superaría `TOTAL_TIME`, ajustamos `dt` para que la simulación termine exactamente en el tiempo deseado. Luego llamamos a `apply_breach`, que recibe los arreglos `h`, `u`, `v`, `zb`, las dimensiones de la malla, el tamaño de las celdas, el tiempo actual, la posición de la presa, el ancho de la brecha, el tiempo de inicio y el tiempo de formación. Esta función verifica si el tiempo actual ya superó el tiempo de inicio, y si es así, comienza a reducir la altura de la presa en las celdas de la brecha de manera gradual, desde el momento de inicio hasta que se completa el tiempo de formación.

A continuación, llamamos a `solver_pass`. Esta es la función más importante, porque es la que da un paso de tiempo a las ecuaciones de aguas someras. Por ahora solo diremos que recibe los arreglos de profundidad y velocidad actuales y nuevos, el arreglo de topografía, las dimensiones de la malla, los tamaños de celda, el paso de tiempo `dt` y el coeficiente de Manning, y que luego, cuando analicemos `utils.c`, la desglosaremos en detalle. Después de `solver_pass`, tenemos los nuevos valores de `h`, `u` y `v` en los arreglos `h_new`, `u_new` y `v_new`. Los copiamos a los arreglos actuales `h`, `u` y `v` usando un bucle `for` que recorre todas las celdas. Este bucle tiene la directiva `#pragma omp parallel for`, que le dice al compilador que reparta las iteraciones entre los núcleos del procesador, porque la copia de 312,000 elementos es una operación que se puede hacer en paralelo sin problemas.

Luego actualizamos el tiempo actual sumándole `dt`, e incrementamos el contador de pasos. Cada vez que la diferencia entre el tiempo actual y el último guardado supera los 30 segundos, llamamos a `save_snapshot`, esta función recibe el arreglo `h`, el número de paso, las dimensiones de la malla, y la ruta donde guardar el archivo. Lo que hace es escribir el arreglo `h` en un archivo binario, para que después podamos leerlo con ***Python*** y generar las visualizaciones. También mostramos en pantalla un mensaje con el tiempo actual, el número de paso, el `dt` que se usó, y los valores máximos y mínimos de `h`, `u` y `v`, para poder monitorear la simulación en tiempo real. Cada 500 pasos, hacemos una verificación de estabilidad: recorremos todas las celdas buscando algún `NaN` (Not a Number) o algún valor de `h` mayor a 1000 metros, que es físicamente imposible. Si encontramos alguno, mostramos un error y salimos del bucle anticipadamente. Esto es una válvula de escape para que el programa no siga ejecutándose si algo salió mal.

Cuando el bucle termina, ya sea porque alcanzamos el tiempo total o porque explotó, medimos el tiempo real que tardó la simulación usando `omp_get_wtime`, que devuelve el tiempo en segundos desde un punto de referencia. Luego mostramos un resumen final: la cantidad de pasos ejecutados, el tiempo simulado, el volumen inicial, el volumen final, y la diferencia porcentual entre ellos, que es una medida de qué tan bien el método conserva la masa. Finalmente, guardamos un último snapshot con los valores finales y liberamos la memoria que reservamos al principio, llamando a `free` para cada uno de los arreglos. Esto es importante porque si no liberamos la memoria, quedaría ocupada hasta que el programa termine, y en simulaciones largas eso podría ser un problema.

Ahora hablemos de la función `calculate_dt_adaptive`, que está definida después de `main`. Esta función recibe los arreglos `h`, `u`, `v`, las dimensiones de la malla `nx` y `ny`, el tamaño de celda `dx` y el factor `cfl`. Lo primero que hace es declarar una variable `max_speed` e inicializarla en cero. Luego, con un bucle `for` que recorre todas las celdas, calcula para cada celda la velocidad de onda como `sqrt(G * fmax(h[i], 0.05)) + sqrt(u[i]*u[i] + v[i]*v[i])`. Usamos `fmax` para asegurarnos de que `h` no sea cero o negativo, porque la raíz cuadrada de un número negativo no está definida. Si la velocidad de onda de esa celda es mayor que `max_speed`, actualizamos `max_speed`. Este bucle tiene la directiva `#pragma omp parallel for reduction(max:max_speed)`, que le dice al compilador que recorra las celdas en paralelo, y que al final combine los máximos parciales de cada núcleo en un solo valor. Si no usáramos esta directiva, un solo núcleo tendría que recorrer las 312,000 celdas solo, y la función sería más lenta.

Una vez que tenemos la velocidad máxima, calculamos `dt` como `cfl * dx / (max_speed + 0.001)`, donde sumamos 0.001 para evitar la división por cero en caso de que `max_speed` sea cero. Finalmente, devolvemos el mínimo entre ese valor y 0.5, para evitar pasos de tiempo demasiado grandes en caso de que el cálculo dé un número muy alto.

Muy bien, ahora que entendemos qué hace el código, podemos **analizar su complejidad algorítmica**, que corresponde a una medida de cuánto trabajo tiene que hacer la computadora en función del tamaño del problema. El tamaño del problema está determinado principalmente por la cantidad de celdas, que llamaremos $N = NX \times NY$. En nuestro caso, $N = 600 \times 520 = 312,000$ celdas. Dentro de cada paso de tiempo, la función `solver_pass` recorre todas las celdas interiores, que son prácticamente todas, y para cada celda realiza varias operaciones aritméticas. La copia de los arreglos también recorre todas las celdas, y la función `calculate_dt_adaptive` también las recorre. Por lo tanto, el **trabajo por paso de tiempo** es proporcional a $N$. Si el número total de pasos de tiempo es $M$, entonces el **trabajo total** es proporcional a $M \times N$.

Ahora, ¿cuánto es $M$? Depende del paso de tiempo `dt`. Si `dt` fuera constante, entonces $M = \text{TOTAL \ TIME} / \text{dt}$. Pero `dt` no es constante porque el método es adaptativo y se ajusta según la velocidad del agua. En nuestra simulación, con `CFL_NUMBER = 0.08`, `dt` suele estar alrededor de 0.05 segundos. Para `TOTAL_TIME = 600` segundos, eso da aproximadamente $M = 600 / 0.05 = 12,000$ pasos. Multiplicado por $N = 312,000$ celdas, obtenemos alrededor de **3.7 mil millones de operaciones elementales en específico**. Ese número es **enorme**, representa un esfuerzo computacional gigante, y por eso la **paralelización** es fundamental.

Cuando usamos las directivas de OpenMP, esas 3.7 mil millones de operaciones se **reparten entre los núcleos del procesador**. Si tenemos ocho núcleos, cada núcleo hace aproximadamente una octava parte del trabajo, es decir, unos 462 millones de operaciones. El tiempo de ejecución se reduce aproximadamente en un factor igual a la **cantidad de núcleos**, aunque en la práctica la aceleración es un poco menor porque hay partes del código que no se pueden paralelizar, como las reducciones y las **escrituras en áreas** compartidas de memoria.

Si no usáramos OpenMP, es decir, si ejecutáramos el programa de forma **serial**, todas esas 3.7 mil millones de operaciones las tendría que hacer un solo núcleo. En una computadora de escritorio típica, eso podría tomar varios minutos o incluso horas. En las máquinas de alta gama de la universidad, que tienen 16 o 32 núcleos, la diferencia es aún más dramática: lo que con paralelización tarda 1 minuto, en versión serial podría tardar más de una hora. Por eso la **paralelización** no es un lujo, es una **necesidad** para problemas de este tamaño.

En términos de **complejidad espacial**, es mucho más fácil. Reservamos siete arreglos de tamaño $N$, cada uno de tipo `double`, que ocupa **8 bytes**. También hay algunos arreglos temporales dentro de las funciones auxiliares, pero son pequeños y no dependen de $N$. Por lo tanto, la memoria total es aproximadamente $7 \times N \times 8 = 56 \times N$ bytes. Con $N = 312,000$, eso es unos **17.5 megabytes**. Es una cantidad razonable que cabe holgadamente en cualquier computadora moderna. Si quisiéramos simular un dominio más grande, por ejemplo con el doble de celdas en cada dirección, $N$ se multiplicaría por cuatro, y la memoria aumentaría a unos 70 megabytes, que todavía es manejable. El verdadero cuello de botella no es la memoria, sino el tiempo de cómputo.

En resumen, el código que acabamos de describir es un ejemplo de cómo se implementa un solver numérico paralelizado en C. Las constantes están todas al principio para facilitar los cambios, la memoria se reserva dinámicamente y se libera al final, el bucle temporal avanza el tiempo con un paso adaptativo que garantiza la estabilidad, y las operaciones costosas se paralelizan con OpenMP para aprovechar los múltiples núcleos del procesador. Sin embargo, existen funciones utilizadas allí, que no definimos directamente en esa fracción de código, por ejemplo ¿Cómo está implementada `solver_pass()`? ¿Cómo se implementó `calculate_volume()`? Esto es importante para entrar mucho más en detalle en la construcción del modelo, observemos la implementación de cada una:

````c
#include <stdio.h>
#include <stdlib.h>
#include <math.h>
#include <omp.h>
#include "utils.h"

#define G 9.81

double calculate_volume(double *h, int n, double dx, double dy) {
    double volume = 0.0;
    #pragma omp parallel for reduction(+:volume)
    for (int i = 0; i < n; i++) {
        volume += h[i];
    }
    return volume * dx * dy;
}

void save_snapshot(double *h, int step, int nx, int ny, const char *folder) {
    char filename[256];
    snprintf(filename, sizeof(filename), "%s/h_%06d.bin", folder, step);
    FILE *f = fopen(filename, "wb");
    if (f) {
        fwrite(h, sizeof(double), nx * ny, f);
        fclose(f);
    }
}

void charge_real_topography(double *zb, int nx, int ny, const char *filename) {
    FILE *f = fopen(filename, "rb");
    if (!f) {
        printf("ERROR: No se pudo abrir %s\n", filename);
        return;
    }
    size_t elements = fread(zb, sizeof(double), nx * ny, f);
    fclose(f);
    if (elements == nx * ny) {
        printf("Topografía real cargada: %d x %d\n", nx, ny);
    } else {
        printf("ERROR: Solo se leyeron %zu de %d elementos\n", elements, nx * ny);
    }
}

void apply_breach(double *h, double *u, double *v, double *zb, int nx, int ny, double dx, 
                  double dy, double time, int dam_x, double breach_width, double breach_start, 
                  double breach_time) {
    if (time < breach_start) return;
    
    double factor = 1.0;
    if (time < breach_start + breach_time) {
        factor = (time - breach_start) / breach_time;
    }
    
    int cells_breach = (int)(breach_width / dx);
    if (cells_breach < 1) cells_breach = 1;
    
    int start_breach = (ny / 2) - (cells_breach / 2);
    int end_breach = start_breach + cells_breach;
    
    int j, idx;
    #pragma omp parallel for private(j, idx)
    for (j = start_breach; j < end_breach; j++) {
        if (j >= ny) continue;
        idx = dam_x + j * nx;
        if (zb[idx] > 50.0) {
            zb[idx] = 50.0 + (zb[idx] - 50.0) * (1.0 - factor);
        }
    }
}

void solver_pass(double *h, double *u, double *v, double *h_new, double *u_new, double *v_new,
                 double *zb, int nx, int ny, double dx, double dy, double dt, double n_manning) {
    int total = nx * ny;
    
    #pragma omp parallel for
    for (int i = 0; i < total; i++) {
        h_new[i] = h[i];
        u_new[i] = u[i];
        v_new[i] = v[i];
    }
    
    #pragma omp parallel for collapse(2)
    for (int j = 1; j < ny - 1; j++) {
        for (int i = 1; i < nx - 1; i++) {
            int idx = i + j * nx;
            
            double h_lim = fmax(h[idx], 0.05);
            double h_left = fmax(h[(i-1) + j*nx], 0.05);
            double h_right = fmax(h[(i+1) + j*nx], 0.05);
            double h_bottom = fmax(h[i + (j-1)*nx], 0.05);
            double h_top = fmax(h[i + (j+1)*nx], 0.05);
            
            double hu_left = h_left * u[(i-1)+j*nx];
            double hu_right = h_lim * u[idx];
            double dhu_dx = (hu_right - hu_left) / dx;
            
            double hv_bottom = h_bottom * v[i+(j-1)*nx];
            double hv_top = h_lim * v[idx];
            double dhv_dy = (hv_top - hv_bottom) / dy;
            
            double new_h = h_lim - dt * (dhu_dx + dhv_dy);
            
            if (new_h < 0.05) new_h = 0.05;
            if (new_h > 200.0) new_h = 200.0;
            h_new[idx] = new_h;
            
            double eta_center = h_lim + zb[idx];
            double eta_left = h_left + zb[(i-1)+j*nx];
            double eta_right = h_right + zb[(i+1)+j*nx];
            double eta_bottom = h_bottom + zb[i+(j-1)*nx];
            double eta_top = h_top + zb[i+(j+1)*nx];
            
            double deta_dx = (eta_right - eta_left) / (2.0 * dx);
            double deta_dy = (eta_top - eta_bottom) / (2.0 * dy);
            
            double max_grad = 0.15;
            if (fabs(deta_dx) > max_grad) deta_dx = (deta_dx > 0) ? max_grad : -max_grad;
            if (fabs(deta_dy) > max_grad) deta_dy = (deta_dy > 0) ? max_grad : -max_grad;
            
            double u_new_val = u[idx] - dt * G * deta_dx;
            double v_new_val = v[idx] - dt * G * deta_dy;
            
            double max_vel = 30.0;
            if (u_new_val > max_vel) u_new_val = max_vel;
            if (u_new_val < -max_vel) u_new_val = -max_vel;
            if (v_new_val > max_vel) v_new_val = max_vel;
            if (v_new_val < -max_vel) v_new_val = -max_vel;
            
            double speed = sqrt(u_new_val * u_new_val + v_new_val * v_new_val);
            if (speed > 0.1 && new_h > 0.1) {
                double Sf = (n_manning * n_manning * speed * speed) / pow(new_h, 4.0/3.0);
                double fric_factor = dt * G * Sf / (speed + 0.1);
                if (fric_factor > 0.5) fric_factor = 0.5;
                u_new_val = u_new_val * (1.0 - fric_factor);
                v_new_val = v_new_val * (1.0 - fric_factor);
            }
            
            u_new[idx] = u_new_val;
            v_new[idx] = v_new_val;
        }
    }
    
    #pragma omp parallel for
    for (int j = 0; j < ny; j++) {
        int idx = 0 + j * nx;
        h_new[idx] = h_new[1 + j * nx];
        u_new[idx] = 0.0;
        v_new[idx] = v_new[1 + j * nx];
    }
    
    #pragma omp parallel for
    for (int j = 0; j < ny; j++) {
        int idx = (nx-1) + j * nx;
        h_new[idx] = h_new[(nx-2) + j * nx];
        u_new[idx] = 0.0;
        v_new[idx] = v_new[(nx-2) + j * nx];
    }
    
    #pragma omp parallel for
    for (int i = 0; i < nx; i++) {
        int idx = i + 0 * nx;
        h_new[idx] = h_new[i + 1 * nx];
        u_new[idx] = u_new[i + 1 * nx];
        v_new[idx] = 0.0;
    }
    
    #pragma omp parallel for
    for (int i = 0; i < nx; i++) {
        int idx = i + (ny-1) * nx;
        h_new[idx] = h_new[i + (ny-2) * nx];
        u_new[idx] = u_new[i + (ny-2) * nx];
        v_new[idx] = 0.0;
    }
}
````

Ahora, en este código, donde encontramos las **funciones auxiliares** reales, es en donde realmente intentamos resolver las ecuaciones, aquí intentamos **modificar la topografía** cuando la presa se rompe, guardamos **los resultados** en archivos y aplicamos las **condiciones de borde**. Vamos a desmenuzarlo con la misma paciencia de antes:

La primera función que encontramos es `calculate_volume`. Recibe un arreglo `h` que contiene la profundidad del agua en cada celda, la cantidad total de celdas `n`, y el tamaño de cada celda `dx` y `dy`. Lo que hace es sumar todas las profundidades y luego multiplicar por el área de una celda para obtener el volumen total de agua en el dominio. La línea importante acá es `#pragma omp parallel for reduction(+:volume)`. Esta directiva le dice al compilador que el bucle `for` debe ejecutarse en paralelo, es decir, que las iteraciones se repartan entre los núcleos del procesador. Pero hay un problema: varias iteraciones están actualizando la misma variable `volume` al mismo tiempo. Si no hacemos nada especial, dos núcleos podrían leer el mismo valor, sumarle el suyo, y escribir el resultado, y se perdería información. La cláusula `reduction(+:volume)` resuelve esto: le dice al compilador que haga una copia local de `volume` en cada núcleo, que cada núcleo sume sus valores en su copia local, y que al final del bucle combine todas las copias sumándolas en la variable original. Sin esta directiva, el cálculo sería incorrecto y obtendríamos un número aleatorio. Con ella, podemos sumar 312 mil números en paralelo de forma segura.

La siguiente función es `save_snapshot`. Recibe el arreglo `h`, el número de paso `step`, las dimensiones de la malla `nx` y `ny`, y una carpeta `folder` donde guardar el archivo. Lo que hace es construir un nombre de archivo con `snprintf`, que es una versión segura de `printf` que escribe en una cadena de texto en lugar de en la pantalla. El nombre tiene la forma `h_000001.bin`, donde los ceros se rellenan para que todos los nombres tengan la misma longitud y sea más fácil ordenarlos después. Luego abre el archivo en modo binario con `fopen`, escribe el arreglo completo con `fwrite`, y lo cierra. Esta función no tiene paralelización porque escribir un archivo es una operación que debe hacerse de forma secuencial; si varios núcleos intentaran escribir el mismo archivo al mismo tiempo, se generarían conflictos.

Luego viene `charge_real_topography`, que es la función que carga la topografía real desde un archivo binario. Recibe el arreglo `zb` donde se guardarán las elevaciones, las dimensiones de la malla, y el nombre del archivo. Abre el archivo, lee todo el contenido de una sola vez con `fread`, y lo guarda en `zb`. Si el archivo no existe, muestra un error y sale. Si la cantidad de elementos leídos no coincide con la cantidad esperada, también muestra un error. Esta función tampoco se paraleliza porque la lectura de archivos es una operación que el sistema operativo maneja de forma secuencial.

La función `apply_breach` es la que modela la rotura de la presa. Recibe los arreglos `h`, `u`, `v`, `zb`, las dimensiones de la malla, el tamaño de las celdas, el tiempo actual `time`, la posición de la presa `dam_x`, el ancho de la brecha `breach_width`, el tiempo de inicio de la rotura `breach_start` y el tiempo que tarda en formarse `breach_time`. Lo primero que hace es verificar si ya pasó el tiempo de inicio. Si no, la función no hace nada y la presa sigue intacta. Si ya pasó, calcula un `factor` que va de 0 a 1 durante el tiempo de formación. Cuando el tiempo está entre `breach_start` y `breach_start + breach_time`, el factor es la fracción de tiempo transcurrido; después de ese tiempo, el factor es 1. Luego calcula cuántas celdas ocupa la brecha dividiendo `breach_width` por `dx`, y determina las filas en las que se va a abrir la brecha, centradas en la mitad del dominio (`ny / 2`). Finalmente, un bucle paralelizado recorre esas filas y, para cada una, calcula el índice de la celda en la posición de la presa (`dam_x + j * nx`) y reduce la elevación de la presa en esa celda. La fórmula `zb[idx] = 50.0 + (zb[idx] - 50.0) * (1.0 - factor)` hace que la presa pase de su altura original (que está por encima de 50 metros) a 50 metros a medida que `factor` aumenta. La directiva `#pragma omp parallel for private(j, idx)` le dice al compilador que paralelice este bucle. Las variables `j` e `idx` son privadas, es decir, cada núcleo tiene su propia copia, y no se comparten entre los núcleos. Esto es necesario porque si varios núcleos modificaran la misma variable al mismo tiempo, los valores se corromperían.

Ahora llegamos al corazón del archivo, donde realmente ocurre la simulación: la función `solver_pass`. Esta es la función que da un paso de tiempo a las ecuaciones de aguas someras. Es larga y compleja, pero vamos a recorrerla con calma. Primero, copia los valores actuales de `h`, `u` y `v` a los arreglos `h_new`, `u_new` y `v_new`. Esta copia se hace con un bucle paralelizado simple, usando `#pragma omp parallel for`. La copia es independiente para cada celda, por lo que se puede hacer en paralelo sin problemas.

Luego viene un bucle doble que recorre todas las celdas interiores de la malla, desde `i = 1` hasta `nx - 2`, y desde `j = 1` hasta `ny - 2`. Este bucle tiene la directiva `#pragma omp parallel for collapse(2)`. La cláusula `collapse(2)` es importante porque le dice al compilador que tome los dos bucles anidados y los trate como un solo bucle gigante para paralelizarlo. Esto es más eficiente que paralelizar solo el bucle externo, porque permite repartir el trabajo de manera más equilibrada entre los núcleos. Sin `collapse(2)`, el compilador paralelizaría solo el bucle de `j`, y cada núcleo se encargaría de una fila completa de `i`. Con `collapse(2)`, las iteraciones se reparten de forma más fina: un núcleo puede hacer una celda aquí y otra allá, y la carga de trabajo se distribuye mejor. Dentro de este bucle, en cada celda, lo primero que hacemos es tomar la profundidad `h[idx]` y la de sus cuatro vecinos, pero asegurándonos de que no sean menores a 0.05 metros con la función `fmax`. Esto es para evitar divisiones por cero o raíces cuadradas de números negativos más adelante. Este valor de 0.05 metros es un límite numérico, no físico; significa que en las celdas que se secan, dejamos una película de agua de 5 centímetros para mantener la estabilidad del método.

Luego calculamos los flujos en la dirección **x** y en la dirección **y**. El flujo en **x** es `hu_left = h_left * u[(i-1)+j*nx]` y `hu_right = h_lim * u[idx]`. Estos son los caudales por unidad de ancho que entran por la cara izquierda y salen por la cara derecha de la celda. La diferencia `(hu_right - hu_left) / dx` es la divergencia del flujo, es decir, cuánto más agua sale de la que entra. Si esta diferencia es positiva, está saliendo más agua de la que entra, y la profundidad debería disminuir. Si es negativa, está entrando más de la que sale, y la profundidad debería aumentar. La ecuación de continuidad discretizada queda como `new_h = h_lim - dt * (dhu_dx + dhv_dy)`, que es la forma en que actualizamos la profundidad.

A continuación, limitamos `new_h` para que no sea menor a 0.05 metros ni mayor a 200 metros. Este límite de 200 metros es una protección numérica. En la realidad, la profundidad del agua en un embalse como Urrá no supera los 73 metros, pero en zonas de la simulación pueden aparecer picos espurios si el método se vuelve inestable, y este límite los controla. Después, calculamos la superficie libre `eta = h + zb`. Esta es la elevación total del agua sobre el nivel de referencia. La pendiente de esta superficie libre en las direcciones **x** e **y** es la que genera la aceleración del agua. Calculamos `deta_dx` y `deta_dy` usando diferencias centradas: `(eta_right - eta_left) / (2 * dx)`. Si la superficie está más alta a la izquierda que a la derecha, `deta_dx` es negativo, y eso empuja el agua hacia la derecha.

Sin embargo, ejecutando la simulación, encontramos un problema: en terrenos muy accidentados, estas **pendientes** pueden ser enormes y causar **aceleraciones** gigantes que vuelven **inestable** el método. Por eso limitamos `deta_dx` y `deta_dy` a un valor máximo de 0.15, que es una pendiente del 15%. Esto es una corrección numérica, no física, pero nos permite mantener la estabilidad cuando la topografía real tiene pendientes muy pronunciadas.

Luego actualizamos las velocidades: `u_new_val = u[idx] - dt * G * deta_dx` y `v_new_val = v[idx] - dt * G * deta_dy`. La aceleración es proporcional a la pendiente de la superficie libre y a la gravedad. Cuanto más inclinada está la superficie, **más se acelera el agua**. También limitamos las velocidades a un máximo de 30 metros por segundo en valor absoluto. Esto es otra protección numérica. En una inundación real, las velocidades pueden superar los 30 m/s en situaciones extremas, pero al limitarlas evitamos que el método explote.



A continuación, aplicamos la fricción de Manning. Calculamos la rapidez `speed = sqrt(u_new_val^2 + v_new_val^2)`. Si la rapidez es mayor a 0.1 m/s y la profundidad es mayor a 0.1 metros, calculamos la pendiente de fricción `Sf = (n^2 * speed^2) / h^{4/3}`. Esta es una fórmula empírica que relaciona la pérdida de energía por fricción con la velocidad y la profundidad. Luego calculamos un factor de fricción `fric_factor = dt * G * Sf / speed` y lo limitamos a 0.5 para que la fricción no frene el agua más de un 50% en un solo paso de tiempo. Finalmente, multiplicamos las velocidades por `(1 - fric_factor)`, lo que las reduce gradualmente.

Después de terminar con todas las celdas interiores, aplicamos las condiciones de borde. Los bordes de la malla se tratan aparte porque no tienen todos los vecinos necesarios para aplicar las fórmulas anteriores. Usamos condiciones reflectivas: en los bordes izquierdo y derecho, la profundidad se copia de la celda adyacente, la velocidad normal se anula (porque el agua no puede atravesar la pared), y la velocidad tangencial se copia. En los bordes inferior y superior, hacemos lo mismo. Estas condiciones simulan que el valle está confinado por montañas y que el agua no puede salirse del dominio por los costados. Cada uno de estos bucles de borde también está paralelizado con `#pragma omp parallel for`, porque son independientes y se pueden repartir entre los núcleos.

Ahora bien, hemos terminado de describir el código que construímos. Ahora vale la pena detenerse un momento y reflexionar sobre lo que hemos hecho: **¿qué virtudes podemos destacar de lo ejecutado?** Y también, muy importante **¿qué limitaciones existen y son inminentes?**, porque de eso se trata, en muchas ocasiones, la ciencia: no es solo entender **lo que funciona**, sino también por qué y en qué **condiciones falla**. En primer lugar, hemos implementado un **solver numérico paralelizado** para las ecuaciones de aguas someras en dos dimensiones. Las ecuaciones que resolvemos son un sistema de tres ecuaciones diferenciales parciales que describen la conservación de la masa,

$$ \frac{\partial h}{\partial t} + \frac{\partial (h u)}{\partial x} + \frac{\partial (h v)}{\partial y} = 0, $$

y la conservación de la cantidad de movimiento en las dos direcciones,

$$ \frac{\partial (h u)}{\partial t} + \frac{\partial}{\partial x}\left(h u^2 + \frac{g h^2}{2}\right) + \frac{\partial (h u v)}{\partial y} = -g h \frac{\partial z_b}{\partial x} - \frac{\tau_{bx}}{\rho}, $$

$$ \frac{\partial (h v)}{\partial t} + \frac{\partial (h u v)}{\partial x} + \frac{\partial}{\partial y}\left(h v^2 + \frac{g h^2}{2}\right) = -g h \frac{\partial z_b}{\partial y} - \frac{\tau_{by}}{\rho}. $$

Estas ecuaciones capturan la **física esencial** de un flujo en aguas **poco profundas**. Hemos logrado discretizarlas utilizando el **método de diferencias finitas**, que es el más intuitivo y el más fácil de paralelizar porque cada celda solo necesita información de sus vecinos inmediatos.

En segundo lugar, hemos aprovechado al máximo las capacidades de **OpenMP** para paralelizar las partes más costosas del código. Los bucles que recorren las 312,000 celdas de la malla se ejecutan ahora en múltiples núcleos simultáneamente. La directiva `#pragma omp parallel for` es la herramienta fundamental: le dice al compilador que divida las iteraciones del bucle entre los hilos de ejecución. En funciones como `calculate_volume`, donde necesitamos combinar resultados parciales, usamos `reduction` para que cada hilo acumule su propia suma local y luego las combine sin errores. En los bucles anidados de `solver_pass`, usamos `collapse(2)` para paralelizar las dos dimensiones juntas y lograr un mejor balance de carga. Esta paralelización no es un lujo; es una necesidad. La cantidad de operaciones que realiza una simulación es del orden de decenas de miles de millones, y sin paralelización, estos cómputos tomarían horas o incluso días en una computadora de escritorio. Con OpenMP, los reducimos a minutos.

En tercer lugar, hemos implementado un **manejo de memoria contiguo y eficiente**. Los arreglos son unidimensionales y se acceden mediante índices planos `idx = i + j * nx`. Esto es mucho más rápido que usar arreglos bidimensionales anidados, porque los datos están almacenados de forma continua en la memoria RAM, lo que mejora la localidad de los datos y reduce los fallos de caché. Y, finalmente, en cuarto lugar, hemos entendido la necesidad, y hemos incorporado **mecanismos de estabilidad** que, aunque limitados, permiten que el código corra sin explotar. El paso de tiempo adaptativo basado en la condición de ***Courant-Friedrichs-Lewy***, los límites de profundidad y velocidad, la limitación de los gradientes de superficie libre, y el factor de fricción acotado, todo ello contribuye a que la simulación pueda avanzar durante cientos de segundos simulados sin generar valores no numéricos.

## **Consideraciones a Futuro Para Lograr el Objetivo y Mejorar el Modelo**

Para nosotros, más que en una entrega final, este proyecto lo vemos como una **oportunidad de crecimiento** en términos técnicos, porque representa un reto **en capacidad de abstracción matemática** y **modelación**, por eso, debemos encontrar los puntos exactos en los que fallamos en este producto, para mejorarlo a futuro, como un proyecto personal para nosotros. 

En nuestra elaboración actual, no es que el programa esté **mal escrito**, es que el método numérico falla y no se adecúa al problema que **queremos resolver**; entonces, **¿qué es lo que nos falta?**, vamos a verlo con calma, porque en cada uno de estos puntos, tenemos una **lección a aprender**.

La primera y más importante limitación de nuestro solver es que **no estamos planteando las ecuaciones completas de Saint-Venant**. En nuestra implementación, la ecuación de continuidad está bien planteada:

$$ \frac{\partial h}{\partial t} + \frac{\partial (h u)}{\partial x} + \frac{\partial (h v)}{\partial y} = 0 $$

Pero simplificamos demasiado la ecuación de **cantidad de movimiento** que consideramos inicialmente. Nos quedamos con una versión reducida:

$$ \frac{\partial u}{\partial t} = -g \frac{\partial \eta}{\partial x} $$
$$ \frac{\partial v}{\partial t} = -g \frac{\partial \eta}{\partial y} $$

Cuando en realidad, la forma completa debería incluir los términos de **advección** y **presión hidrostática**. Para la dirección **x**, la ecuación correcta es:

$$ \frac{\partial (h u)}{\partial t} + \frac{\partial}{\partial x}\left(h u^2 + \frac{g h^2}{2}\right) + \frac{\partial (h u v)}{\partial y} = -g h \frac{\partial z_b}{\partial x} - \frac{\tau_{bx}}{\rho} $$

El término $\frac{g h^2}{2}$ dentro de la derivada espacial es la **presión hidrostática**, y es fundamental para que el agua se "empuje" a sí misma. Sin él, las ondas no se propagan correctamente. Los términos $h u^2$ y $h u v$ son la **advección del momento**, y son los responsables de que la cantidad de movimiento se transporte con el flujo. Sin ellos, la velocidad en un punto solo cambia por la pendiente local de la superficie libre, pero no porque llegue agua con mayor velocidad desde aguas arriba.

**¿Qué tendríamos que hacer?** En una próxima implementación, debemos trabajar con las **variables conservativas** $$ \mathbf{U} = (h, h u, h v) $$ en lugar de las variables primitivas $(h, u, v)$. Así, la actualización de cada celda se haría sobre el vector conservativo, y los flujos entre celdas se calcularían resolviendo un **problema de Riemann** en cada cara. Eso nos permitiría capturar correctamente las ondas de choque y las ondas de rarefacción que caracterizan una rotura de presa.

Luego, n nuestro código, manejamos las zonas donde el agua es muy poco profunda con un truco brutal: forzamos una profundidad mínima de 5 centímetros en todas las celdas, así estén en medio de una montaña. Esa lámina ficticia de agua no es física; es un parche numérico para evitar divisiones por cero. Pero tiene consecuencias graves: pues sin tenerlo en cuenta, estabamos **creando masa artificial**, **generando velocidades de onda donde no debería haberlas**, y **abriendo canales numéricos** por donde el flujo puede escapar o meterse sin sentido físico.

En las visualizaciones, esto se traduce en ese ruido de alta frecuencia que vemos en las zonas que deberían estar secas: pequeñas oscilaciones que no tienen nada que ver con la realidad. Es como si el agua estuviera "temblando" en las laderas de las montañas.


Luego, también debemos ver que introdujimos **límites artificiales para la profundidad** ($h \leq 200$ m) y para las velocidades ($|u|, |v| \leq 30$ m/s). En teoría, deberían ser medidas de seguridad que solo actúan en situaciones extremas. En la práctica, las velocidades están pegadas a **30 m/s durante casi toda la simulación**, y las profundidades tocan los 200 m con frecuencia. Eso significa que **los clamps no están corrigiendo excepciones; están definiendo la solución**. El modelo quiere que las variables sean mucho más grandes, pero nosotros se lo impedimos.

Esto es inaceptable para una simulación seria, y se debe corregir en base a las correcciones anteriores.

Luego, en `apply_breach`, modelamos la rotura de la presa reduciendo gradualmente la elevación del terreno en las celdas de la brecha:

```
zb[idx] = 50.0 + (zb[idx] - 50.0) * (1.0 - factor);
```

Esto hace que la presa desaparezca, pero no genera la **discontinuidad hidráulica** que caracteriza una rotura real. En la física de una rotura de presa, lo que ocurre es que de repente se abre una compuerta y el agua embalsada, que estaba en reposo, comienza a caer por un desnivel abrupto. Eso genera una onda de choque que viaja aguas abajo y una onda de rarefacción que viaja aguas arriba. Nuestra simple modificación de la topografía no captura esa dinámica.

En lugar de modificar `zb`, deberíamos imponer una **condición de borde interna** en la posición de la presa durante el tiempo de formación de la brecha. Por ejemplo, podríamos calcular el caudal que sale por la brecha usando la fórmula de vertedero:

$$ Q = C_d \cdot L \cdot (H_{upstream} - H_{brecha})^{3/2} $$

donde $C_d$ es un coeficiente de descarga, $L$ es el ancho de la brecha, y $H_{brecha}$ es la elevación de la cresta de la brecha en cada instante. Ese caudal se inyecta en las celdas aguas abajo como una condición de contorno, mientras que el nivel del embalse aguas arriba se actualiza según el volumen que se va vaciando.

Además, cuando cargamos la topografía real del DEM de Urrá, los gradientes de la superficie libre se volvieron enormes. Nuestro esquema, que no está **bien balanceado**, no pudo manejar el equilibrio entre el término de presión y el término de pendiente del fondo. En un esquema **bien balanceado**, la solución de **"lago en reposo"** (superficie libre constante, velocidad cero) se mantiene exactamente, independientemente de lo accidentada que sea la topografía. El nuestro no lo hace, porque calculamos la aceleración como $-g \partial \eta / \partial x$, donde $\eta = h + z_b$. Si $\eta$ es constante, la aceleración es cero, y eso está bien. El problema es que en la discretización, los errores de truncamiento pueden generar pequeñas aceleraciones espurias que, acumuladas, producen flujos artificiales. Eso es exactamente lo que vimos.

Para solucionarlo, debemos implementar un esquema **de buen balance** para el término fuente de la pendiente del fondo. Una forma común es discretizar el término $-g h \partial z_b / \partial x$ de manera que se cancele exactamente con el gradiente de presión hidrostática en condiciones de reposo. Esto se logra usando una **reconstrucción de la superficie libre** en las caras de las celdas, y evaluando la pendiente del fondo de forma consistente con esa reconstrucción.



Cambiar cada uno de estos **puntos** es un trabajo difícil, casi un proyecto en sí mismo, sin embargo, sabemos cuál es el camino a seguir, rehaciendo el núcleo para que no sea una simplificación tan grande de las ecuaciones que **queremos resolver**, realizando estas correcciones, lograremos un resultado mucho mejor, aprovechando las prácticas de paralelización funcionales que **ya tenemos** en nuestra implementación.
